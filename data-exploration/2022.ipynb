```python
# in df, remove 'P-' from patientId only from the start of the string
df['patientId'] = df['patientId'].str.lstrip('P-')
df.head()
```

```python
# load cll_broad_2015_clinical_data into another dataframe
# in df['patientId'], remove the 'P-' prefix
# then join the two dataframes on 'patientId' and show how many rows are in the joined dataframe
df_2015 = pd.read_csv("../data/cbioportal/cll_broad_2015_clinical_data.csv")
df_2015['patientId'] = df_2015['patientId'].str.lstrip('CLL-')
df_2015 = df_2015.rename(columns={'patientId': 'patientId_2015'})
display(df_2015.head())
```

```python
df[df['patientId'] == 'GCLL-0182']
```

```python
df_2015[df_2015['patientId_2015'] == 'GCLL-0182']
```

```python
# perform a left join on the two dataframes based on 'patientId' and 'patientId_2015'
merged_df = pd.merge(df, df_2015, left_on='patientId', right_on='patientId_2015', how='left')
# display the shape of the merged dataframe
print(merged_df.shape)
# display the first few rows of the merged dataframe
display(merged_df.head())
```

```python
row = merged_df[merged_df['patientId'] == 'GCLL-0182']
display(row)
```

```python
# find how many rows in df have a matching patientId in df_2015
matching_rows = df[df['patientId'].isin(df_2015['patientId_2015'])]
print(f"Number of matching rows: {matching_rows.shape[0]}")
# write non matchin patientIds to a file
non_matching_patient_ids = df[~df['patientId'].isin(df_2015['patientId_2015'])]['patientId']
with open('./non_matching_patient_ids.json', 'w') as f:
    json.dump(non_matching_patient_ids.tolist(), f, indent=4)
```

```python
# load ids from non_matching_patient_ids.json and make new dataframe with those patientIds
with open('./non_matching_patient_ids.json', 'r') as f:
    non_matching_patient_ids = json.load(f)
non_matching_df = df[df['patientId'].isin(non_matching_patient_ids)]
# display the first few rows of the non_matching_df
display(non_matching_df.head(), non_matching_df.shape)
```
